```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
```

```python

# 1. Load your dataset
df = pd.read_csv("cleaned_india.csv")

# 2. Load a specialized Bangla NLP model 
# This model is explicitly trained to understand Bengali semantics
print("Loading Bangla sentence transformer...")
model = SentenceTransformer('shihab17/bangla-sentence-transformer')

# 3. Convert headlines into numerical vectors (Embeddings)
# We cluster based on headlines because they contain the core "event" with less noise
print("Encoding headlines...")
embeddings = model.encode(df['News_Content'].tolist())

# 4. Group similar events together using Agglomerative Clustering
print("Clustering events...")
clustering = AgglomerativeClustering(
    n_clusters=None, 
    distance_threshold=0.35, # Tweak this: lower = more specific events, higher = broader events
    metric='cosine',        # Measures the angle between text vectors
    linkage='average'       # Required when using cosine distance
)

df['event_cluster'] = clustering.fit_predict(embeddings)

# 5. Save the results
df.to_csv("clustered_india.csv", index=False)
print(f"Success! Grouped 100 articles into {df['event_cluster'].nunique()} distinct flash events.")
```
