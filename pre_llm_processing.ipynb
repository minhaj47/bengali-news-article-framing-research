```python
import pandas as pd
import json
import uuid
import sys
```

```python
def convert_csv_to_jsonl(input_csv_path: str, output_jsonl_path: str):
    """
    Transforms a raw CSV file containing Bengali text into a robust JSONL format.
    Ensures perfect UTF-8 encoding and assigns UUIDs for tracking.
    """
    try:
        # Read the CSV file, explicitly defining UTF-8 to prevent mojibake
        df = pd.read_csv(input_csv_path, encoding='utf-8')
          
        # Verify required columns exist in the raw data
        required_columns = ['News_Title', 'News_Content']
        for col in required_columns:
            if col not in df.columns:
                raise ValueError(f"Missing required column: {col}")
                  
        processed_count = 0
          
        with open(output_jsonl_path, 'w', encoding='utf-8') as outfile:
            for index, row in df.iterrows():
                # Generate a unique identifier for downstream relational mapping
                article_id = str(uuid.uuid4())
                  
                # Clean and strip trailing whitespaces or erroneous line breaks
                record = {
                    "article_id": article_id,
                    "title": str(row['News_Title']).strip(),
                    "content": str(row['News_Content']).strip(),
                }
                  
                # ensure_ascii=False is mandatory to preserve native Bengali script
                json_string = json.dumps(record, ensure_ascii=False)
                outfile.write(json_string + '\n')
                processed_count += 1
                  
        print(f"Successfully processed {processed_count} articles into {output_jsonl_path}")
          
    except Exception as e:
        print(f"Error processing dataset: {str(e)}")
        sys.exit(1)

if __name__ == "__main__":
    # Point to your raw dataset file
    convert_csv_to_jsonl('secular.csv', 'llm_secular.jsonl')
```
